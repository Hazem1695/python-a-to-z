# **1️⃣ What is a Module?**

A module is simply a Python file (.py) that contains code.

Example:
```
math_utils.py
```
Inside the file:
```
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b
```
Now this file is a module.

***Why modules are important***

| Benefit           | Explanation                    |
| ----------------- | ------------------------------ |
| Code Organization | split large programs           |
| Reuse             | use code in many projects      |
| Maintainability   | easy debugging                 |
| Collaboration     | teams work on separate modules |

Example structure:
```
project/
    main.py
    data_processing.py
    model.py
    visualization.py
```
This is modular programming.


# **2️⃣ Importing Modules**

Python provides several ways.

***1️⃣ Basic Import***

````python
import math

print(math.sqrt(16))
````

***2️⃣ Import Specific Functions***

````python
from math import sqrt

print(sqrt(25))
````

***3️⃣ Import Multiple Items***

````python
from math import sqrt, pi

print(sqrt(16))
print(pi)
````

***4️⃣ Import Everything (Not Recommended)***

````python
from math import *

print(pi)
````

Problem:

⚠️ Namespace conflicts.

***5️⃣ Alias Import***

Example with NumPy:

````python
import numpy as np

arr = np.array([1,2,3])
print(arr)
````

# **4️⃣ Most Important Built-in Modules**

These are the most useful modules you must know.

````python
import math

math.sqrt(25)       # Returns the square root of 25 → 5.0
math.pow(2, 3)      # Raises 2 to the power of 3 → 8.0
math.factorial(5)   # Calculates 5! (5 × 4 × 3 × 2 × 1) → 120
math.pi             # Mathematical constant π (approx. 3.14159)
math.e              # Mathematical constant e (Euler’s number, approx. 2.71828)
math.ceil(4.2)      # Rounds 4.2 up to the nearest integer → 5
math.floor(4.9)     # Rounds 4.9 down to the nearest integer → 4
````

| Function    | Description |
| ----------- | ----------- |
| sqrt()      | square root |
| pow()       | power       |
| factorial() | factorial   |
| ceil()      | round up    |
| floor()     | round down  |


# *random module*

Generate random numbers.

```
import random
```

***Random integer***

````python
import random

random.randint(1,10)    # Generates a random integer between 1 and 10 (inclusive).
````

````python
import random

# Generate 5 random numbers between 1 and 10
for i in range(5):
    print(random.randint(1, 10))
````

***Random float***

````python
import random

random.random()   # Returns a random floating-point number between 0.0 and 1.0 (inclusive of 0.0, exclusive of 1.0).
````

````python
import random

# Generate 5 random floating-point numbers between 0 and 1
for i in range(5):
    print(random.random())
````

***Choose random element***

````python
import random

random.choice(["CS", "AI", "IT"])   # Randomly selects one element from the given list.
````

````python
import random

# Run multiple times to see different results
for i in range(5):
    print(random.choice(["CS", "AI", "IT"]))
````

***Shuffle list***

````python
import random

data = [1, 2, 3, 4]
random.shuffle(data)   # Shuffles the list in place, rearranging the elements randomly.
print(data)
````

# **datetime module**

Work with dates and time.

````python
import datetime

now = datetime.datetime.now()   # Gets the current local date and time from your system clock.
print(now)                      # Prints the full timestamp, e.g. year-month-day hour:min:sec
````

***Create date***

````python
from datetime import date

today = date.today()   # Gets the current local date from your system clock.
print(today)           # Prints today’s date in YYYY-MM-DD format, e.g. year-month-day
````

# **json module**

Work with JSON data.



***Convert dictionary → JSON***

````python
import json

data = {"name": "Hazem", "age": 30}
json_data = json.dumps(data)   # Converts the Python dictionary into a JSON-formatted string.
print(json_data)               # Output: {"name": "Hazem", "age": 30}
````

***Convert JSON → dictionary***

````python
import json

json.loads(json_data)   # Converts a JSON-formatted string back into a Python object (usually a dict).
````

# **5️⃣ Creating Your Own Module**

Create file:
Use the `%%writefile` magic command to create a new Python file




````python
%%writefile calculator.py

def add(a, b):
    return a + b

def sub(a, b):
    return a - b
````

***Use the module***

````python
import calculator

print(calculator.add(5,3))

print(calculator.sub(10, 4))
````

# **6️⃣ __name__ Special Variable**

Important for modules.

Example:

````python
print(__name__)
````

Output if file executed directly:
```
__main__
```

Why we use it

````python
def main():
    print("Program running")

if __name__ == "__main__":
    main()
````

Meaning:

Run this code only if file executed directly.

# **7️⃣ Packages**

A package is a folder containing modules.

Structure:
```

    utils/

        file_utils.py
```

***Import from package***
```

```

```

```

# **9️⃣ Absolute vs Relative Imports**

Absolute Import (Best)
```
from utils.math_utils import add
```
Relative Import
```
from .math_utils import add
```
`.` means current package

# **1️⃣2️⃣ Virtual Environments**

VERY IMPORTANT for real projects.

Why?

Different projects need different package versions.

Create virtual environment
```
python -m venv venv
```
***Activate***

Windows
```
venv\Scripts\activate
```
Mac/Linux
```
source venv/bin/activate
```
***Install packages inside environment***
```
pip install numpy
```
Now only this project uses it.

***Deactivate***
```
deactivate
```
