# **1️⃣ Lists**
**What is a List?**

A list is an ordered, mutable collection of elements.

```
Ordered → elements have position
Mutable → elements can change
```
***Use List when***

- order matters

- you need to modify data

- store sequences

Example:
```
numbers = [10, 20, 30, 40]
```
Lists can store different types:
```
data = [1, "Hazem", True, 3.14]
```

Creating Lists

````python
numbers = [1,2,3]
names = ["Ali", "Sara", "Hazem"]
empty = []
````

Using constructor:

````python
numbers = list([1,2,3])
````

Accessing Elements

````python
numbers = [10,20,30,40]

print(numbers[0])
print(numbers[-1])
````

````python
fruits = ["apple", "banana", "orange"]
print(fruits[2])
print(fruits[-3])
````

List Slicing

````python
numbers = [1,2,3,4,5]
print(numbers[1:4])
````

reverse()

copy()

Copy list.

````python
numbers = [3,1,4,2]
new_list = numbers.copy()
print(new_list)
````

````python
data = [1,2,3]
copy_data = data.copy()
data.clear()
print(data)
print(copy_data)
````

List Comprehensions

````python
squares = [x**2 for x in range (5)]
print(squares)
````

Nested Lists

`matrix[index of list][index of element on the list]`

# **2️⃣ Tuples**
What is a Tuple?

A tuple is an ordered but immutable collection.
```
Ordered
Immutable (cannot change)
```
***Use Tuple when***

- data should not change

- representing coordinates

- returning multiple values

Example:
```
point = (10,20)
```

Creating Tuples

````python
t = (1,2,3)
t = tuple([1,2,3])
````

Single element tuple:

````python
t = (5,)
````

⚠️ comma is required.

Accessing Elements

````python
t = (10,20,30)
print(t[0])
print(t[1])
print(t[2])
````

Tuple Methods

Tuples have only two methods.

count()

````python
t = (1,2,2,3)
print(t.count(2))
````

````python
fruits = ("Apple", "Banana", "Orange", "Apple")
print(fruits.count("Apple"))
````

index()

````python
t = (10,20,30)
print(t.index(20))
````

````python
fruits = ("Apple", "Banana", "Orange")
print(fruits.index("Apple"))
````

Tuple Unpacking

Very common in Python.

````python
point =(10,20)
x, y = point
print(x)
print(y)
````

````python
fruits = ("Apple", "Banana", "Orange")
x, y, z = fruits
print(x)
print(y)
print(z)
````

Built-in Functions

````python
numbers = (4, 7, 2, 9, 5)

print(len(numbers))
print(max(numbers))
print(min(numbers))
print(sum(numbers))
````

# **3️⃣ Sets**
What is a Set?

A set is:
```
unordered
unique elements
mutable
```
***Use Set when***

- removing duplicates

- checking membership quickly

Example:
```
numbers = {1,2,3}
```
Duplicates automatically removed.


````python
numbers = {1,1,2,3}
print(numbers)
````

Creating Sets

````python
s = {1,2,3}
s = set([1,2,3])
````

Empty set:

````python
s = set()
````

Set Methods

add()

````python
s = {1,2}
s.add(3)
print(s)
````

````python
fruits = {"Apple", "Banana", "Orange"}
fruits.add("Watermelon")
print(fruits)
````

update()

Add multiple elements.

````python
s = {1,2}
s.update([3,4,5])
print(s)
````

````python
fruits = {"Apple", "Banana", "Orange"}
fruits.update(["Watermelon","Juava"])
print(fruits)
````

remove()

````python
s = {1,2}
s.remove(2)
print(s)
````

````python
fruits = {"Apple", "Banana", "Orange"}
fruits.remove("Apple")
print(fruits)
````

⚠️ Error if element not exist.

No error.

pop()

Removes random element.

````python
s = {1,2,3,4}
s.pop()
print(s)
````

````python
fruits = {"Apple", "Banana", "Orange"}
fruits.pop()
print(fruits)
````

clear()

````python
s = {1,2,3,4}
s.clear()
print(s)
````

````python
fruits = {"Apple", "Banana", "Orange"}
fruits.clear()
print(fruits)
````

Set Operations (VERY IMPORTANT)

| Operation                | Mathematical Meaning                | Python Operator | Python Method            | Example   | Result                       |    |                                    |
| ------------------------ | ----------------------------------- | --------------- | ------------------------ | --------- | ---------------------------- | -- | ---------------------------------- |
| **Union**                | Elements in **either set**          | `               |                        | `union()` | `A | B` | All unique elements from both sets |
| **Intersection**         | Elements **common to both sets**    | `&`             | `intersection()`         | `A & B`   | Elements that appear in both |    |                                    |
| **Difference**           | Elements in **A but not in B**      | `-`             | `difference()`           | `A - B`   | Elements only in A           |    |                                    |
| **Symmetric Difference** | Elements in **A or B but not both** | `^`             | `symmetric_difference()` | `A ^ B`   | Elements unique to each set  |    |                                    |


union (|)

A	| B`	All unique elements from both sets

````python
a = {1,2}
b = {2,3}
print(a | b)
````

````python
fruits1 = {"Apple", "Banana", "Orange"}
fruits2 = {"Watermelon","Juava","Apple"}
print(fruits1 | fruits2)
````

intersection (&)

Elements that appear in both

````python
a = {1,2}
b = {2,3}
print(a & b)
````

````python
fruits1 = {"Apple", "Banana", "Orange"}
fruits2 = {"Watermelon","Juava","Apple"}
print(fruits1 & fruits2)
````

difference (-)

Elements in A but not in B

````python
a = {1,2}
b = {2,3}
print(a - b)
````

````python
fruits1 = {"Apple", "Banana", "Orange"}
fruits2 = {"Watermelon","Juava","Apple"}
print(fruits1 - fruits2)
````

symmetric difference (^)

Elements in A or B but not both

````python
a = {1,2}
b = {2,3}
print(a ^ b)
````

````python
fruits1 = {"Apple", "Banana", "Orange"}
fruits2 = {"Watermelon","Juava","Apple"}
print(fruits1 ^ fruits2)
````

membership

````python
s = {1,2,3,4,5,6}
if 3 in s:
  print("3 is in the set")
````

Set Comprehensions

````python
squares = {x**2 for x in range(5)}
print(squares)
````

Removing Duplicates

````python
items = [1, 2, 2, 3, 4, 4, 5]
unique_items = set(items)
print(unique_items)
````

# **4️⃣ Dictionaries**
What is a Dictionary?

A dictionary stores key-value pairs.

```
student = {
    "key": "value",
    "key": value,
    "key": "value"
}
```
***Use Dictionary when***

- mapping information

- structured data

- JSON-like objects
Example:

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}
````

Access Values

````python
print(student["name"])
````

Dictionary Methods

keys()

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.keys()
````

values()

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.values()
````

items()

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.items()
````

get()

Safer access.

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.get("name")
````

update()

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.update({"age":26})
print(student.get("age"))
print(student)
````

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.pop("age")
print(student)
````

popitem()

Removes last item.

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.popitem()
print(student)
````

clear()

Remove everything.

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

student.clear()
print(student)
````

copy()

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

new_dict = student.copy()
student.pop("name")
print(f"student dict: {student}")
print(f"new_dict: {new_dict}")
````

Set Default

Adds key if missing, otherwise returns existing value

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}
student.setdefault("grade", "A")

print(student.items())
````

Looping Through Dictionary

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

for key,value in student.items():
  print(f"{key}:", value)
````

Dictionary Comprehension


````python
squares = {x: x**2 for x in range(5)}
print(squares)
````

Membership Test

````python
student = {
    "name": "Hazem",
    "age": 20,
    "major": "Computer science"
}

if "name" in student:
    print("Key exists")
````

Nested Dictionaries


````python
students = {
    "Hazem": {
        "age":24,
        "grade":"A"
    },
    "Ali": {
        "age": 22,
        "grade": "B"
    }
}

print(students["Hazem"]["grade"])
````

# **5️⃣ collections Module**

The collections module provides advanced data structures in Python.

Important ones:
```
Counter
defaultdict
deque
```
These are part of Python’s standard library under the collections.

**Counter**

Used for counting elements.

````python
from collections import Counter

data = ["a","b","a","c","b","a"]

count = Counter(data)
print(count)
````

Most common element:

````python
from collections import Counter

data = ["a","b","a","c","b","a"]

count.most_common(1)
````

defaultdict

Automatically creates default values.

Example:

````python
from collections import defaultdict

d = defaultdict(int)

d["a"] +=1

print(d)
````

deque

A double-ended queue.

Very fast for insert/ remove.


Methods:
```
append()
appendleft()
pop()
popleft()
```
Example:

````python
from collections import deque

d = deque([1,2,3])
d.append(5)
print(d)
````

````python
from collections import deque

d = deque([1,2,3])
d.appendleft(5)
print(d)
````

````python
from collections import deque

d = deque([1,2,3])
d.pop()
print(d)
````

````python
from collections import deque

d = deque([1,2,3])
d.popleft()
print(d)
````

# **When to Use Each Structure**

| Structure   | Use case          |
| ----------- | ----------------- |
| List        | ordered data      |
| Tuple       | fixed data        |
| Set         | unique values     |
| Dict        | key-value mapping |
| Counter     | counting          |
| defaultdict | missing keys      |
| deque       | fast queue        |


# **Python Data Structures Comparison**

| Feature              | List                 | Tuple        | Set                 | Dictionary               |
| -------------------- | -------------------- | ------------ | ------------------- | ------------------------ |
| Syntax               | `[ ]`                | `( )`        | `{ }`               | `{key: value}`           |
| Ordered              | ✅ Yes                | ✅ Yes        | ❌ No                | ✅ Yes (Python 3.7+)      |
| Mutable (can change) | ✅ Yes                | ❌ No         | ✅ Yes               | ✅ Yes                    |
| Allows duplicates    | ✅ Yes                | ✅ Yes        | ❌ No                | ❌ Keys: No (Values: Yes) |
| Indexing             | ✅ Yes                | ✅ Yes        | ❌ No                | ❌ Use keys instead       |
| Key–Value structure  | ❌ No                 | ❌ No         | ❌ No                | ✅ Yes                    |
| Use case             | General collections  | Fixed data   | Unique elements     | Mapping data             |
| Example              | `[1,2,3]`            | `(1,2,3)`    | `{1,2,3}`           | `{"name":"Hazem"}`       |
| Typical operations   | append, remove, sort | count, index | union, intersection | keys, values, items      |


# **Quick Memory Trick 🧠**

Think like this:

| Structure  | Think of it as               |
| ---------- | ---------------------------- |
| List       | **Dynamic array**            |
| Tuple      | **Read-only list**           |
| Set        | **Unique bag of items**      |
| Dictionary | **Hash table (key → value)** |


# **Complexity Insight (Important for Algorithms)**

| Operation | List | Set  | Dict |
| --------- | ---- | ---- | ---- |
| Search    | O(n) | O(1) | O(1) |
| Insert    | O(1) | O(1) | O(1) |
| Delete    | O(n) | O(1) | O(1) |
