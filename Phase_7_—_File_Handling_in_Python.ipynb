# **1️⃣ Reading Files in Python**

What is File Reading?

Reading a file means loading the content of a file into your Python program.

Example files:

.`txt`

.`csv`

.`json`

.`log`

.`xml`

Basic Syntax
```
file = open("filename.txt", "r")
content = file.read()
file.close()
print(content)
```
⚠️Note: if the file is not in the same folder with notebook you should take file path like this

path in Windows
```
file = open("C:\\Users\\User\\Documents\\data.txt", "r")
```
⚠️ Use double backslashes `\\` or raw strings
```
file = open(r"C:\Users\User\Documents\data.txt", "r")

```
path in Linux/Mac
```
file = open("/home/user/Documents/data.txt", "r")

```

````python
file = open("data1.txt", "r")
content = file.read()
file.close()
print(content)

file.close()
````

| Part    | Meaning           |
| ------- | ----------------- |
| open()  | open the file     |
| "r"     | read mode         |
| read()  | read file content |
| close() | close the file    |

# **3️⃣ Reading Methods**

Python gives several methods.

1️⃣ `read()`

Reads all file content

````python
file = open("data1.txt", "r")
print(file.read())
file.close()
````

2️⃣ readline()

Reads one line

````python
file = open("data1.txt", "r")
line = file.readline()
print(line)
file.close()
````

3️⃣ readlines()

Reads all lines as list

````python
file = open("data1.txt", "r")
lines = file.readlines()
print(lines)
file.close()
````

4️⃣ Loop Through File

Very common in data processing

````python
file = open("data1.txt", "r")

for line in file:
  print(line)

file.close()
````

5️⃣ Best Practice ***(with statement)***

Instead of closing manually.

Use context manager.

````python
with open("data1.txt", "r") as file:
  content = file.read()
  print(content)
````

Advantages:

✅ Automatically closes file

✅ Cleaner code

✅ Prevents memory issues

This is the professional way.


6️⃣ Writing Files

Writing means saving data into a file.


Write Mode (w)

⚠️ Important

`w` overwrites the file

Example:

````python
with open("data1.txt","w") as file:
  file.write("Hello World\n")
  file.write("Python is amazing")

with open("data1.txt","r") as file:
  content = file.read()
  print(content)
````

***Append Mode (a)***

Append means add new content without deleting old data.

````python
with open("data1.txt", "a") as file:
  file.write("\nNew line added")

with open("data1.txt","r") as file:
  content = file.read()
  print(content)
````

# **7️⃣ CSV Files**

CSV = Comma Separated Values

CSV is used everywhere in:

- Data science

- Machine learning

- Excel datasets

⚠️ Note: This code creates CSV file that will be used later in this notebook.

````python
import csv

# data
data = [
    ["Hazem", 29, "Cairo"],
    ["Ali", 30, "Giza"],
    ["Ahmed", 22, "Alex"]
]

# create csv file
with open("People data.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # write column names
    writer.writerow(["name", "age", "city"])

    # write rows
    writer.writerows(data)

print("CSV file created successfully!")
````

Explanation

| Code                 | Meaning                     |
| -------------------- | --------------------------- |
| `import csv`         | Import CSV library          |
| `csv.writer(file)`   | Create writer object        |
| `writer.writerow()`  | Write one row               |
| `writer.writerows()` | Write multiple rows         |
| `newline=""`         | Prevents empty lines in CSV |


***Reading CSV***

Python has a built-in module called csv

````python
import csv

with open("People data.csv", "r") as file:
  reader = csv.reader(file)

  for row in reader:
    print(row)
````

Writing CSV

````python
import csv

data = [
    ["name", "age", "city"],
    ["Akram", 25, "Cairo"],
    ["Anas", 36, "Alex"]
]

with open("People data.csv", "w", newline="") as file:
  writer = csv.writer(file)
  writer.writerows(data)

with open("People data.csv", "r") as file:
  reader = csv.reader(file)

  for row in reader:
    print(row)
````

# **8️⃣ JSON Files**

JSON = JavaScript Object Notation

Used everywhere:

- APIs

- Web

- AI datasets

- configuration files

⚠️ Note: This code creates Json file that will be used later in this notebook.

````python
import json

# data
data = {
    "name": "Anas",
    "age": 36,
    "city": "Cairo"
}

# create json file
with open("person.json", "w") as file:
    json.dump(data, file, indent=4)

print("JSON file created successfully!")
````

Reading JSON

````python
import json

with open("person.json", "r") as file:
  data = json.load(file)

print(data)
````

Access values:

````python
print(data["name"])
print(data["city"])
````

Writing JSON

````python
import json

data = {
    "name" : "Akram",
    "age" : 25,
    "city" : "Cairo"
}

with open("data.json", "w") as file:
  json.dump(data, file)


with open("person.json", "r") as file:
  data = json.load(file)

print(data)
````
