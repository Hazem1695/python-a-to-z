# **Phase 8 — Error Handling in Python**

real programs always face errors (wrong input, missing files, API errors, etc.). Good developers handle errors instead of letting the program crash.

# **1️⃣ What is an Error?**

An error is a problem that occurs during program execution.

Example:


````python
print(10 / 0)
````

The program crashes ❌.

Error handling allows us to control this behavior.

# **2️⃣ Types of Errors in Python**

| Type          | Description                      | Example          |
| ------------- | -------------------------------- | ---------------- |
| Syntax Error  | Wrong Python syntax              | missing colon    |
| Runtime Error | Happens during execution         | division by zero |
| Logical Error | Program runs but result is wrong | wrong formula    |

***Example of Syntax Error***


````python
if 5 > 2
    print("Hello")
# you forget to add (:) at the end of condition
````

***Example of Runtime Error***

````python
x = int("hello")
# it take integer value
````

# **3️⃣ Try / Except**

The most important structure.

***Syntax***
```
try:
    # risky code
except:
    # code if error happens
```
Example


````python
try:
  x = int(input("Enter number: "))
  print(10 / x)

except:
  print("Error happend!")
````

If user enters 0 or text, program will not crash

# **4️⃣ Catch Specific Errors**

Better practice is catching specific errors.

Example:

````python
try:
  x = int(input("Enter number: "))
  result = 10 / x

except ZeroDivisionError:
  print("You cannot divide by zero!")
except ValueError:
  print("Invalid number!")
````

````python
try:
  x = int(input("Enter number: "))
  result = 10 / x

except ZeroDivisionError:
  print("You cannot divide by zero!")
except ValueError:
  print("Invalid number!")
````

# **5️⃣ Multiple Exceptions**

You can catch several errors.

````python
try:
  num = int(input("Enter number: "))
  print(10 / num)
except (ZeroDivisionError, ValueError):
  print("Invalid number!")
````

# **6️⃣ Using else**

`else` runs only if no error occurs.

````python
try:
  x = int(input("Enter number: "))
  result = 10 / x

except ZeroDivisionError:
  print("Cannot divide by zero")

else:
  print("Result: ", result)
````

Flow:
```
try -> success -> else
try -> error -> except
```

````python
try:
  x = int(input("Enter number: "))
  result = 10 / x

except ZeroDivisionError:
  print("Cannot divide by zero")

else:
  print("Result: ", result)
````

# **7️⃣ Using finally**

`finally` always runs whether error occurs or not.

Example:

````python
try:
  file = open("data.txt")

except FileNotFoundError:
  print("File not fount")

finally:
  print("Program finished")
````

Used often for:

- closing files

- closing database connections

- cleanup tasks

# **8️⃣ Full Structure**

Professional structure:
```
try:
    # risky code

except ErrorType:
    # handle error

else:
    # run if success

finally:
    # always execute
```


# **9️⃣ Raising Exceptions**

You can create errors manually using raise.

Example:

````python
age = int(input("Enter age: "))

if age < 0:
  raise ValueError("Age cannot be negative")
````

Used when input is invalid.

# **🔟 Custom Exceptions**

You can create your own error types.

Example:

````python
class InvalidAgeError(Exception):
  pass
````

Use it:

````python
age = int(input("Enter age: "))

if age < 18:
  raise InvalidAgeError("Age must be at least 18")
````

This is used in large systems and frameworks

# **1️⃣1️⃣ Example: File Handling + Error Handling**



````python
try:
  with open("data.txt","r") as file:
    content = file.read()
    print(content)

except FileNotFoundError:
  print("File does not exist")

except PermissionError:
  print("No permission to read the file")
````

# **1️⃣2️⃣ Example: API / Data Input**

This pattern is very common in AI pipelines.

````python
try:
    number = int(input("Enter number: "))
    result = 100 / number
    print("Result:", result)

except ValueError:
    print("Please enter a valid number")

except ZeroDivisionError:
    print("Number cannot be zero")

# general exception handler: to prevent unexpected error happens (like a file issue, or a keyboard interrupt)
except Exception as e:
    print("An unexpected error occurred:", e)
````
