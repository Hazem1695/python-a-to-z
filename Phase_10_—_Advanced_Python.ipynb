# **1️⃣ Iterators**
**What is an Iterator?**

An iterator is an object that allows you to traverse elements one at a time.

It implements two methods:

| Method       | Purpose                 |
| ------------ | ----------------------- |
| `__iter__()` | returns iterator object |
| `__next__()` | returns next value      |


***Basic Example***

````python
numbers = [1,2,3,4]

iterator = iter(numbers)    # Create an iterator object from the list

print(next(iterator))
print(next(iterator))
print(next(iterator))
print(next(iterator))
````

***How `for` loop works internally***
```
for n in numbers:
    print(n)
```
Internally Python does:



````python
numbers = [1, 2, 3, 4]       # A list of integers

iterator = iter(numbers)     # Create an iterator object from the list

while True:                  # Infinite loop
  try:
    value = next(iterator)   # Get the next item from the iterator
    print(value)             # Print the item
  except StopIteration:      # Raised when no items are left
    break                    # Exit the loop
````

***Creating Custom Iterator***

Example: countdown iterator.

````python
class Countdown:

  def __init__(self, start):
    self.start = start        # Initialize with a starting number

  def __iter__(self):
    return self               # The iterator object is the class itself

  def __next__(self):
    if self.start <= 0:       # Stop when we reach 0
      raise StopIteration     # Signal that iteration is finished

    current = self.start      # Save the current value
    self.start -= 1           # Decrease by 1 for the next call
    return current            # Return the current value
````

1.   `__iter__()` returns the iterator object itself. That’s why you can use `for ... in ....`

2.    `__next__()` defines how to get the next value. Each call reduces `self.start` until it reaches 0.

3.   When `self.start <= 0`, StopIteration is raised, which tells Python the loop is done.





````python
counter = Countdown(5)

for num in counter:
  print (num)
````

`*Another Example: Even Numbers Iterator*`

````python
class EvenNumbers:

    def __init__(self, max_num):
        self.max = max_num       # Upper limit for even numbers
        self.current = 0         # Start from 0

    def __iter__(self):
        return self              # The iterator object is the class itself

    def __next__(self):
        if self.current > self.max:   # Stop when exceeding max_num
            raise StopIteration

        value = self.current     # Save the current even number
        self.current += 2        # Move to the next even number
        return value             # Return the current value
````

1.   `self.current` starts at 0 and increases by 2 each time.
2.   Iteration stops once `self.current` exceeds `max_num`.
3.   This is a stateful iterator: it remembers where it left off between calls to `next()`.









````python
evens = EvenNumbers(10)

for n in evens:
    print(n)
````

***2️⃣ Generators***

Generators are simpler iterators.

Instead of `__next__`, we use `yield`.

***Simple Generator***

````python
def my_generator():
  yield 1
  yield 2
  yield 3


g = my_generator()
print(next(g))
print(next(g))
print(next(g))
````

***Generator with Loop***

````python
def count_up(n):

  for i in range(n):
    yield i

for num in count_up(5):
  print(num)
````

***Fibonacci Generator***

````python
def Fibonacci(n):
  a, b = 0, 1              # Start with 0 and 1

  for _ in range(n):       # Loop n times
    yield a                # Yield the current value of 'a'
    a, b = b, a + b        # Update values: next Fibonacci pair

for num in Fibonacci(10):
  print(num)
````

1.  `yield` makes this a generator function. Instead of returning all values at once, it produces them one by one, pausing between calls.
2.  Each iteration returns the current Fibonacci number `(a)`, then updates `a` and `b` to the next pair.
3.  The loop runs `n` times, so you get the first `n` Fibonacci numbers

Note: if you don’t actually need the loop variable, you can use _ to signal that it’s intentionally unused.



***Infinite Generator***

````python
def infinite_numbers():
    n = 0
    while True:        # Infinite loop
        yield n        # Produce the current value of n
        n += 1         # Increment n for the next iteration

gen = infinite_numbers()   # Create the generator object

for _ in range(5):
    print(next(gen))       # Get the next value from the generator
````

1. infinite_numbers() defines a generator that will keep yielding numbers starting from 0, forever.
2. gen = infinite_numbers() creates the generator object.
3. The for _ in range(5) loop runs 5 times.

    - Each time, next(gen) is called, which resumes the generator where it left off.

    - The generator yields the next number (0, 1, 2, 3, 4) and pauses until the next call.


**Generator Expressions**

Similar to list comprehension.

````python
# Create a generator that yields squares of numbers from 0 to 9
squares = (x*x for x in range(10))

for s in squares:
    print(s)
````

# **3️⃣ Decorators**

Decorators modify functions without changing them.

They are widely used in frameworks like FastAPI.

***Basic Decorator***

````python
def my_decorator(func):

  def wrapper():
    print("Before function")

    func()

    print("After function")

  return wrapper
````

````python
@my_decorator
def say_hello():
  print("Hello")

say_hello()
````

***Decorator Without @ Syntax***

````python
def say_hi():
  print("hi")

say_hi = my_decorator(say_hi)

say_hi()
````

***Decorator with Arguments***

````python
def decorator(func):
  def wrapper(a, b):

    print("Adding number")

    result = func(a,b)
    print("Result: ", result)
    return result

  return wrapper
````

````python
@decorator
def add(x, y):
  return x + y

add(3,4)
````

***Timing Decorator***

Very useful for performance.

````python
import time

def timer(func):
  def wrapper(*args, **kwargs):

    start = time.time()  # Record the start time before calling the function
    result = func(*args, **kwargs)   # Call the original function with any arguments and store its result
    end = time.time()  ## Record the end time after the function finishes

    print("Execution time: ", end-start)   # Print the execution time by subtracting start from end
    return result

  return wrapper
````

````python
@timer
def slow_function():

    total = 0

    for i in range(1000000):
        total += i

slow_function()
````

***Decorator with Parameters***

````python
# Define a decorator factory that takes an argument 'n'
def repeat(n):

    # This is the actual decorator that will wrap the target function
    def decorator(func):

        # The wrapper adds new behavior around the original function
        def wrapper(*args, **kwargs):
            # Run the function 'n' times
            for _ in range(n):
                func(*args, **kwargs)

        return wrapper  # Return the wrapper so it replaces the original function

    return decorator    # Return the decorator so it can be applied with @repeat(n)
````

````python
@repeat(3)
def greet():
    print("Hello")

greet()
````

# **4️⃣ Context Managers**

Used for resource management.

***Example: file handling.***
```
file = open("data.txt")

data = file.read()

file.close()
```
Problem: if error occurs → file not closed.

***With Context Manager***
```
with open("data.txt") as file:
    data = file.read()
```

File automatically closes.

***Creating Custom Context Manager***

Using class.




````python
# Define a custom context manager class
class MyContext:

    # The __enter__ method is called when the 'with' block starts
    def __enter__(self):
        print("Entering")
        # You can return a resource here if needed (e.g., a file object)
        # return self

    # The __exit__ method is called when the 'with' block ends
    # It receives information about any exception that occurred inside the block
    def __exit__(self, exc_type, exc_val, exc_tb):
        print("Exiting")
        # Returning False (or nothing) means exceptions will propagate
        # Returning True would suppress exceptions
````

````python
with MyContext():
    print("Inside block")
````

***Context Manager using decorator***

````python
from contextlib import contextmanager

@contextmanager
def my_context():
    print("Start")
    yield
    print("End")
````

````python
with my_context():
    print("Inside")
````

# **5️⃣ Multiprocessing**

Used to run multiple processes simultaneously.

Python has GIL limitation, so multiprocessing is important.
```
from multiprocessing import Process
```
***Basic Example***


````python
from multiprocessing import Process

def worker():
  print("Worker running")
````

````python
# Create a Process object, passing the target function to run
p = Process(target=worker)

p.start()  # Start the process (this launches a new Python interpreter process)
p.join()   # Wait for the process to finish before continuing
````

***Multiple Processes***

````python
from multiprocessing import Process

def task(num):
    print("Process", num)

# Create a list to keep track of all process objects
processes = []

# Launch 5 separate processes
for i in range(5):
    # Create a new process, passing 'task' as the target function
    # and 'i' as its argument
    p = Process(target=task, args=(i,))

    # Start the process (runs in parallel with the main program)
    p.start()

    # Add the process object to the list for later tracking
    processes.append(p)

# Wait for all processes to finish
for p in processes:
    p.join()
````

***Using Pool***

````python
from multiprocessing import Pool
def square(x):
    return x * x

# Create a pool of 4 worker processes
with Pool(4) as p:
    # Use 'map' to apply 'square' to each element in the list [1,2,3,4]
    # The work is distributed across the 4 processes
    result = p.map(square, [1, 2, 3, 4])

print(result)
````

# **6️⃣ Asynchronous Programming**

Used for high-performance IO operations.

Libraries like FastAPI use this heavily.

```
import asyncio
```
***Basic async function***

````python
import asyncio

async def hello():
  print("Hello")              # Step 1: Print immediately

  await asyncio.sleep(2)      # Step 2: Pause for 2 seconds without blocking the whole program

  print("World")              # Step 3: Print after the pause


await hello()
````

***Multiple async tasks***

````python
import asyncio

async def task1():
    await asyncio.sleep(1)
    print("Task1 done")

async def task2():
    await asyncio.sleep(2)
    print("Task2 done")

async def main():

    await asyncio.gather(task1(), task2())

await main()
````

***Async Web Requests Example***
```
import asyncio

async def fetch(url):

    async with aiohttp.ClientSession() as session:

        async with session.get(url) as response:
            return await response.text()
```


# **7️⃣ Memory Management**

Python manages memory automatically.

Using Garbage Collector.

***Reference Counting***

````python
import sys

a = []

print(sys.getrefcount(a))
````

1.  `sys.getrefcount(obj)`: Returns the number of references pointing to the object `obj`.
2.  When you call it, Python temporarily creates an extra reference to `a` (because you’re passing `a` into the function).
3.  So the number you see will always be at least 2:
    - One reference from the variable `a`.
    - One temporary reference from the function call itself.


***Garbage Collection***

````python
import gc

gc.collect()
````

1.  `gc.collect()` triggers a full garbage collection cycle.
2.  Python’s garbage collector looks for objects that are no longer referenced anywhere in your program and frees their memory.
3.  It returns the number of objects it found that were unreachable and collected.


***Checking memory objects***

````python
gc.get_objects()
````

`gc.get_objects()` is a function that returns a list of all objects currently tracked by Python’s garbage collector.


***Weak References***

````python
import weakref

class MyClass:
    pass

obj = MyClass()              # Create a normal object
weak = weakref.ref(obj)      # Create a weak reference to 'obj'

print(weak())                # Dereference the weak reference
````

1. `obj` is a normal instance of `MyClass`.
2. `weakref.ref(obj)` creates a weak reference — it points to `obj` but does not increase its reference count.
3. `weak()` returns the actual object if it still exists, otherwise `None`.
