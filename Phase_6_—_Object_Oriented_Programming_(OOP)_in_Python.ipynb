# **7️⃣ Encapsulation**

Encapsulation = protecting internal data.

Python visibility types:

| Type      | Syntax |
| --------- | ------ |
| Public    | name   |
| Protected | _name  |
| Private   | __name |

Example:

````python
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance       # private attribute (double underscore)

account = BankAccount(150)
print(account._balance)
````

Will fail.

Why?

Python renames it internally.
```
_BankAccount__balance
```
Encapsulation prevents direct modification.


````python
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance         # private attribute (double underscore)

account = BankAccount(150)
# this is not the correct way
print(account._BankAccount__balance)
````

- Getter: use to access attribute
- Setter: use to modify attribute

````python
class BankAccount:

    def __init__(self, name, balance):
        self.__name = name
        self.__balance = balance
    # Getter & setter
    def get_name(self):
      return self.__name

    def set_name(self, new_name):
      self.__name = new_name

    def get_balance(self):
      return self.__balance

    def set_balance(self, new_balance):
      self.__balance = new_balance



client1 = BankAccount("Ahmed",200)
print(client1.get_name())
client1.set_name("Ali")
print(client1.get_name())

print(client1.get_balance())
client1.set_balance(100)
print(client1.get_balance())

````

Example:

````python
class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.__balance = balance            # private attribute (double underscore)

  def deposit(self, amount):
    self.__balance += amount

  def get_balance(self):
    return self.__balance

account = BankAccount("Ali", 1000)
account.deposit(500)
print(account.get_balance())
````

📌 `__balance` is private (not meant to be accessed directly).

Encapsulation forces you to use methods like `deposit()` and `get_balance()`.


# **8️⃣ @property**

A Python decorator that lets you access methods like attributes

`@property` allows controlled attribute access.

````python
class BankAccount:

    def __init__(self, name, balance):
        self.__name = name
        self.__balance = balance
    # Getter & setter
    @property
    def get_name(self):
      return self.__name

    def set_name(self, new_name):
      self.__name = new_name

    @property
    def get_balance(self):
      return self.__balance

    def set_balance(self, new_balance):
      self.__balance = new_balance



client1 = BankAccount("Ahmed",200)

print(client1.get_name)
client1.set_name("Ali")
print(client1.get_name)

print(client1.get_balance)
client1.set_balance(100)
print(client1.get_balance)
````

Even though `get_name`, `get_balance` are a methods, we use it like a variable.

# **9️⃣ Inheritance**

Inheritance means one class inherits another.

Parent → Child


````python
# bass class (parent)
class Employee:
  def __init__(self,name, email_address, phone, department, address, year_of_birth):
    self.name = name
    self.email_address = email_address
    self.phone = phone
    self.department = department
    self.address = address
    self.year_of_birth = year_of_birth

    def get_name(self):
      return self.name

    def set_name(self, new_name):
      self.name = new_name

    def get_email_address(self):
      return self.email_address

    def set_email_address(self, new_email_address):
      self.email_address = new_email_address

    def get_phone(self):
      return self.phone

    def set_phone(self, new_phone):
      self.phone = new_phone
    # countine setter and getter if you want

  def work(self):
    print("all employee must work 8 hours")

````

````python
# sub class (child)
class Developer(Employee):

  def __init__(self,name, email_address, phone, department, address, year_of_birth, project_name):
    super().__init__(name, email_address, phone, department, address, year_of_birth)
    self.project_name = project_name

  def get_project_name(self):
    return self.__project_name

  def set_project_name(self, new_project_name):
    self.__project_name = new_project_name

   # To print object
  def __str__(self):
    return (f"name: {self.name}, emailAddress: {self.email_address}, "
            f"phone: {self.phone}, department: {self.department}, "
            f"address: {self.address}, yearOfBirth: {self.year_of_birth}, "
            f"projectName: {self.project_name}")

  # OR

  def display(self):
        print("name:", self.name)
        print("emailAddress:", self.email_address)
        print("phone:", self.phone)
        print("department:", self.department)
        print("address:", self.address)
        print("yearOfBirth:", self.year_of_birth)
        print("projectName:", self.project_name)



````

````python
e1 = Developer ("Ali", "ali@yahoo.com", 15536, "IT", "Egypt", "1/1/1995", "OOP")

e1.work()
print("(display function)")
e1.display()
print("(str method)")
print(e1)
````

Developer gets all Employee features.

# **1️⃣2️⃣ @classmethod**

Class methods work with the class itself.

features:
- take cls parameter
- not take self

use case: when you need to work with the class itself

Example:

````python
class Student:

  school = "MIT"

  @classmethod
  def change_school(cls, name):
    cls.school = name

Student.change_school("Stanford")
````

cls = class reference.

````python
class Student:
    def __init__(self, name, year_of_birth):
        self.name = name
        self.year_of_birth = year_of_birth

    @classmethod
    def from_age(cls, name, age):
        year = 2026 - age
        return cls(name, year)

s1 = Student("Ali", 1995)

s2 = Student.from_age("Ahmed", 31)

print(s1.year_of_birth)  # 1995
print(s2.year_of_birth)  # 1995
````

# **1️⃣4️⃣ Magic Methods**

Magic methods start with:
```
__

```
Examples:

| Method   | Purpose     |
| -------- | ----------- |
| **init** | constructor |
| **str**  | printable   |
| **len**  | length      |
| **add**  | addition    |


````python
class Book:
  def __init__(self, title, pages):
    self.title = title
    self.pages = pages

  def __str__(self):
    return f"Books: {self.title}"

  def __len__(self):
    return  self.pages

  def __add__(self, other):
    return self.pages + other.pages

b1 = Book("Python Basics", 200)
b2 = Book("Advanced Python", 300)


print(b1)
print(len(b1))
print(b1 + b2)
````

# **1️⃣5️⃣ Advanced Dunder Methods**

Examples:

| Method      | Example               |
| ----------- | --------------------- |
| **eq**      | ==                    |
| **lt**      | <                     |
| **getitem** | indexing              |
| **call**    | function-like objects |

Example

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = x, y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __getitem__(self, index):
        return (self.x, self.y)[index]

    def __call__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(2, 3)

print(v1 == v2)
print(v1[0])
print(v1())
````

# **1️⃣7️⃣ Abstract Classes**

- Definition: Classes that cannot be instantiated directly. They define a template for subclasses.
- Module: `abc` (Abstract Base Classes).
- Purpose: Force child classes to implement certain methods.


````python
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius ** 2

c = Circle(5)
print(c.area())
````

You cannot do Shape() because it’s abstract

# **1️⃣9️⃣ Multiple Inheritance & MRO**

Multiple inheritance means:
```
Child inherits multiple parents
```
- Multiple Inheritance: A class can inherit from more than one parent.
- MRO: Python decides which parent’s method to call using the C3 linearization algorithm.



````python
class A:
    def greet(self):
        print("Hello from A")

class B:
    def greet(self):
        print("Hello from B")

class C(A, B):
    pass

c = C()
c.greet()  # Hello from A (because A comes first in MRO)
````

Python resolves order using MRO.

Check order:

````python
C.mro()
````

````python
class A:
    def greet(self):
        print("Hello from A")

class B:
    def greet(self):
        print("Hello from B")

class C(A, B):
    def greet(self):
        super().greet()  # Follows MRO

c = C()
c.greet()  # Hello from A
````

# **2️⃣0️⃣ Metaclasses**

What Are Metaclasses?
- In Python, everything is an object — even classes.
- A metaclass is the “class of a class.”
It defines how classes behave, just like a class defines how objects behave.

Analogy
- Objects are built from classes.
- Classes are built from metaclasses.

Hierarchy
```
object
   |
 type (metaclass)
   |
 class
   |
 object instance
```
Default Metaclass

By default, all classes in Python are instances of the built-in `type` metaclass.
```
class MyClass:
    pass

print(type(MyClass))  # <class 'type'>
```

This means `MyClass` itself is created by `type`.


````python
class Meta(type):
    pass


class MyClass(metaclass=Meta):
    pass
````

Here, when Python creates `MyClass`, it calls `MyMeta.__new__`

Used in:

- Django

- ORMs

- frameworks
