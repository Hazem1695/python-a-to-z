# **1️⃣ String Indexing**

Each character in a string has an index (position).

Example:

`name = "Hazem"`

Positions:
```
H   a   z   e   m
0   1   2   3   4
```


Access characters like this:


````python
name = "Hazem"

print(name[0])
print(name[1])
print(name[4])
````

**Negative Indexing**

Python can count from the end.


```
H   a   z   e   m
-5 -4  -3  -2  -1
```



````python
name = "Hazem"
print(name[-1])
print(name[-2])
````

# **2️⃣ String Slicing**

Slicing extracts part of a string.

Syntax


```
string[start:end]
```



````python
name = "Hazem"
print(name[0:3])
````

why the output not Hazi rhater than Haz ?


Because Python stops before index 3.

More examples:

````python
name = "Ahmed"
print(name[:3]) # from start to 2 index
print(name[2:]) # from 2 index to end
print(name[1:4]) # from 1 inex to 3 index
print(name[0:5]) # from 0 index to 4 index
````
