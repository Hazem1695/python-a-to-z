# **2️⃣ The for Loop**

The for loop is used to iterate over a sequence.

Examples of sequences:

- list

- string

- range

- tuple

- dictionary

**Basic Syntax**
```
for variable in sequence:
    code
```
Example 1 — Loop over numbers


````python
for i in range(5):
  print(i)
````

Example 2 — Loop through a string

````python
name = "Hazem"
for letter in name:
  print(letter)
````

Example 3 — Loop through a list

````python
numbers = [10, 20, 30]

for num in numbers:
  print(num)
````

# **4️⃣ The while Loop**

while repeats code as long as a condition is True.

Syntax
```
while condition:
    code
```
Example



````python
i = 1

while i <= 5:
  print(i)
  i += 1
````

````python
i = 1

while i <= 5:
  print(i)
  i = i+ 1
````

Important ⚠️

You must update the variable, otherwise the loop will never stop.

Example of infinite loop:

````python
while True:
    print("Hello")
````

This runs forever.

# **5️⃣ break Statement**

break stops the loop immediately.

Example

````python
for i in range(10):
  if i == 5:
    break
  print(i)
````

Loop stopped when i = 5.

````python
for i in range (1, 10):
  if i == 5:
    break
  print(i)
print("Loop ended")
````

````python
numbers = [2, 4, 6, 8, 10, 12]

for num in numbers:
  if num > 8:
    break
  print(num)
````

````python
names = ["Ali", "Sara", "Hazem", "Omar"]

for n in names:
  if n == "Hazem":
    print("Found Hazem")
    break
````

````python
names = ["Ali", "Sara", "Hazem", "Omar"]

for n in names:
  if n == "Hazem":
    print("Found Hazem")
    break
  print(n)
````

````python
x = 1
while True: # infinite loop
  print(x)
  if x == 3:
    break
  x = x + 1
# break can make (infinite loop) to (normal loop)
````

````python
while True:
    text = input("Enter something (type 'exit' to stop): ")
    if text == "exit":
        print("Loop stopped by user.")
        break
    print("You entered:", text)
````

````python
num = int(input("Enter number of students: "))

for i in range(num):
    grade = int(input(f"Enter grade for student {i+1}: "))

    # Validation
    if grade < 0 or grade > 100:
        print("❌ Invalid grade! Please enter a number between 0 and 100.")
        break

    # Grading system
    if grade >= 90:
        print("Grade: A+")
    elif grade >= 80:
        print("Grade: B+")
    elif grade >= 70:
        print("Grade: B")
    elif grade >= 60:
        print("Grade: C")
    elif grade >= 50:
        print("Grade: D")
    else:
        print("Grade: F")

print("✅ Program finished")
````

# **6️⃣ continue Statement**

continue skips the current iteration and moves to the next.

Example

````python
for i in range (5):
  if i == 2:
    continue
  print(i)
````

Number 2 was skipped.

````python
for i in range (3, 6):
  if i == 3:
    continue
  print(i)
````

The loop skips printing 3

Example 2: Skip negative numbers


````python
numbers = [5, -2, 7, -1, 10]

for num in numbers:
  if num < 0:
    continue
  print(num)
````

Example 3: Skip vowels in a word

````python
word = "Hazem"

for ch in word:
  if ch in "aeiou":
    continue
  print(ch)
````

Example 4: Skip even numbers

````python
for i in range(1, 10):
    if i % 2 == 0:
        continue
    print(i)
````

````python
while True:
    text = input("Enter a word (type 'exit' to stop): ")
    if text == "exit":
        break
    if len(text) < 3:
        print("Too short, skipping...")
        continue
    print("You entered:", text)
````

````python
num = int(input("How many words will you enter? "))

for i in range(num):
    text = input(f"Enter word {i+1} (type 'exit' to stop): ")

    if text == "exit":
        print("Loop stopped by user.")
        break

    if len(text) < 3:
        print("Too short, skipping...")
        continue

    print("You entered:", text)
````

# **7️⃣ pass Statement**

pass means do nothing.

It is used as a placeholder.

Example1: Empty loop

````python
for i in range(5):
  pass
````

This loop runs 5 times but does nothing.

Example for future code:

````python
if True:
    pass
````

This prevents syntax errors when code is not ready yet.

````python
numbers = [1, -2, 3, -4, 5]

for num in numbers:
  if num < 0:
    pass # skip negative numbers for now
  else:
    print(num)
````

# **8️⃣ Nested Loops**

Nested loop = loop inside another loop.

Example:


````python
for i in range (3):   # outer loop
  for j in range (3): #inner loop
    print(i, j)
````

Visualization

Outer loop
```
0
1
2
```
For each value, inner loop runs again.

Example — Multiplication Table

````python
for i in range (1, 6):       #outer loop
  for j in range(1, 6):      #inner loop
    print(i * j, end= " ")   #print product of i and j
  print()                    # move to next line after inner loop

````

`Row 1 → 1*1, 1*2, 1*3, 1*4, 1*5`

`Row 2 → 2*1, 2*2, 2*3, 2*4, 2*5`

The `end` parameter in `print`() controls what is printed at the end of each output.
- `end="\n"` → default (new line after each print).
- `end=" "` → adds a space instead of a new line.
- `end=""` → adds nothing at all (no space, no newline).




Example — Pattern Printing

````python
for i in range(5):
  for j in range (i+1):
    print("*", end=" ")
  print()
````

# **Advance program**

````python
num = int(input("Enter number of students: "))

grades = []  # to store all grades
distribution = {"A+":0, "B+":0, "B":0, "C":0, "D":0, "F":0}  # grade counts

for i in range(num):
    grade = int(input(f"Enter grade for student {i+1}: "))

    # Validation
    if grade < 0 or grade > 100:
        print("❌ Invalid grade! Please enter a number between 0 and 100.")
        break

    grades.append(grade)  # store grade

    # Grading system + count distribution
    if grade >= 90:
        print("Grade: A+")
        distribution["A+"] += 1
    elif grade >= 80:
        print("Grade: B+")
        distribution["B+"] += 1
    elif grade >= 70:
        print("Grade: B")
        distribution["B"] += 1
    elif grade >= 60:
        print("Grade: C")
        distribution["C"] += 1
    elif grade >= 50:
        print("Grade: D")
        distribution["D"] += 1
    else:
        print("Grade: F")
        distribution["F"] += 1

# Show statistics if grades were entered
if grades:
    average = sum(grades) / len(grades)
    print("\n📊 Class Statistics:")
    print(f"Average Grade: {average:.2f}")
    print("Grade Distribution:")
    for grade, count in distribution.items():
        print(f"{grade}: {count}")
else:
    print("No valid grades entered.")
````
