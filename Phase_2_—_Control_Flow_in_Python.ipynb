# **Phase 2 — Control Flow in Python**

Control Flow means:

Controlling the order in which code runs depending on conditions.

Example idea:
```
If age > 18 → allow login
Else → deny access
```

# **1️⃣ The if Statement**

The if statement allows Python to execute code only if a condition is True.

**Basic Syntax**

````python
age = 20
if age >= 18:
  print("You are an adult")
````

# **3️⃣ The else Statement**
``` else ``` runs when the condition is False.

**Syntax**
```

    code

    code
```
Example

````python
age = int(input("Enter your age: "))
if age >= 18:
  print("You are an adult")
else:
  print("You are under 18")
````

# **4️⃣ The elif Statement**

`elif` means else `if`.

It allows multiple conditions.

**Syntax**
```
if condition:
    code
elif condition:
    code
elif condition:
    code
else:
    code
```
Example

````python
score = int(input("Enter youer score: "))

if score >= 90:
  print("Grade A")
elif score >= 80:
  print("Grade B")
elif score >= 70:
  print("Grade C")
else:
  print("Fail")
````

Logical operators combine multiple conditions.
| Operator | Meaning                      |
| -------- | ---------------------------- |
| `and`    | both conditions must be True |

| `not`    | reverses the result          |

Both must be True.

````python
is_logged_in = False

if not is_logged_in:
    print("Please login")
````

# **1️⃣ match-case**

`match-case` in Python (Structural Pattern Matching)

`match-case` allows a program to compare a value against multiple patterns and execute code depending on the match.

Think of it as a cleaner alternative to many `elif` statements.

1️⃣ Basic Syntax

```
match variable:
    case value1:
        code
    case value2:
        code
    case _:
        default_code
```
Important parts:
| Part    | Meaning                    |
| ------- | -------------------------- |
| `match` | value to test              |
| `case`  | pattern to compare         |
| `_`     | default case (like `else`) |


# **2️⃣ Basic Example**

````python
day = input("Enter day: ")

match day:
  case "Saturday":
    print("Weekend")
  case "Sunday":
    print("Weekend")
  case "Monday":
    print("Work day")
  case _:
    print("Normal day")
````

# **3️⃣ Match Numbers**

````python
number = int(input("Enter number: "))

match number:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Other number")
````

# **4️⃣ Multiple Values in One Case**

You can match several values together using `|`.

`|` means OR.

Example:

````python
day = input("Enter day: ")

match day:
  case "Saturday" | "Sunday":
    print("Weekend")
  case "Monday":
    print("Start of week")
  case _:
    print("Regular day")

````

# **5️⃣ Match with Conditions (Guards)**

You can add extra conditions using `if`.

This is very powerful.

````python
number = int(input("Enter number: "))

match number:
  case n if n > 0:
    print("Positive")
  case n if n < 0:
    print("Negative")
  case _:
    print("Zero")
````

# **6️⃣ Matching Lists (Pattern Matching)**

This is where `match` becomes very powerful.

Example:

````python
data = [1,2]

match data:
  case [1,2]:
    print("List contains 1 and 2")
  case [x,y]:
    print("Two elements:", x, y)
  case _:
    print("Something else")
````

````python
x = int(input("enter your x"))
y = int(input("enter your y"))
data = [x,y]
match data:
  case [1,2]:
    print("List contains 1 and 2")
  case [x,y]:
    print("Two elements:", x, y)
  case _:
    print("Something else")
````

# **7️⃣ Capturing Values**

Example:

````python
point = (3, 4)

match point:
    case (x, y):
        print("Point coordinates:", x, y)
````

# **8️⃣ Ignoring Values**

Use _ to ignore values.

````python
point = (5, 10)

match point:
    case (x, _):
        print("x value:", x)
````

# **9️⃣ Practical Example**

Menu system:

````python
choice = input("Choose option (add, delete, update): ")

match choice:
    case "add":
        print("Adding item")

    case "delete":
        print("Deleting item")

    case "update":
        print("Updating item")

    case _:
        print("Invalid option")
````

# **🔟 When Should You Use match-case?**

Use `match-case` when:

✔ many possible values

✔ cleaner than many `elif` statements

✔ pattern matching (lists, tuples, structures)

Example:

Bad:
```
if x == 1:
elif x == 2:
elif x == 3:
elif x == 4:
```
Better:
```
match x:
```
# **⚠️ Important Notes**

1️⃣ Works only in Python 3.10+

2️⃣ _ is the default case

3️⃣ Order matters (first matching case runs)

**Example Combining Everything**

````python
command = input("Command: ")

match command:
  case "start":
    print("Starting system")
  case "stop":
    print("Stopping system")
  case "restart":
    print("Restarting system")
  case "exit":
    print("Goodbye")
  case _:
    print("Unkown command")
````
