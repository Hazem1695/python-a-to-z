# 5️⃣`*args`

`*args` allows functions to accept any number of positional arguments.
```
def add_numbers(*args):
    print(args)
```

````python
def add_numbers(*args):
  print(args)

add_numbers(1,2,3,4,5)
````

It returns a tuple.

Example sum function:

````python
def total(*args):
  result = 0       # Initialize a variable 'result' to store the sum


  for num in args: # Loop through each number passed into the function
    result += num  # Add the current number to 'result'

  return result    # After the loop, return the final sum

print(total(1,2,3,4,5))
````

Valid vs Invalid for `*args `

`*args` collects positional arguments and stores them as a tuple.

| Function Call           | Valid / Invalid    | Explanation                                 |
| ----------------------- | ------------------ | ------------------------------------------- |
| `func(1,2,3)`           | ✅ Valid            | All values go into `args`                   |
| `func("a","b","c")`     | ✅ Valid            | Any type is allowed                         |
| `func(1,[2,3],{"a":5})` | ✅ Valid            | Mixed data types allowed                    |
| `func()`                | ✅ Valid            | `args` becomes empty tuple `()`             |
| `func(a=1,b=2)`         | ❌ Invalid for args | These are keyword arguments, not positional |
| `func(*[1,2,3])`        | ✅ Valid            | List unpacked into positional arguments     |
| `func(**{"a":1})`       | ❌ Invalid for args | `**` is for keyword arguments               |


# 6️⃣ `**kwargs`

`**kwargs` accepts any number of keyword arguments.
```
def show_info(**kwargs):

    print(kwargs)
```

````python
def show_info(**kwargs):
  print(kwargs)

show_info(name="Hazem", age= 25)
````

It returns a dictionary.

````python
def show_info(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}:", value)

show_info(name="Hazem", age=25, field="CS")
````

Valid vs Invalid for `**kwargs`

`**kwargs` collects keyword arguments and stores them as a dictionary.

| Function Call                | Valid / Invalid      | Explanation                     |
| ---------------------------- | -------------------- | ------------------------------- |
| `func(name="Hazem", age=25)` | ✅ Valid              | Stored as dictionary            |
| `func(a=1,b=2,c=3)`          | ✅ Valid              | Multiple keyword arguments      |
| `func()`                     | ✅ Valid              | Empty dictionary `{}`           |
| `func(**{"name":"Hazem"})`   | ✅ Valid              | Dictionary unpacking            |
| `func(1,2,3)`                | ❌ Invalid for kwargs | These are positional arguments  |
| `func(**[1,2,3])`            | ❌ Invalid            | Must unpack a dictionary        |
| `func(**{"a":1,2:3})`        | ❌ Invalid            | Dictionary keys must be strings |


***Valid vs Invalid when combining them***

Correct order in a function definition:

`normal parameters → *args → **kwargs`

| Function Definition            | Valid / Invalid |
| ------------------------------ | --------------- |
| `def func(a, *args, **kwargs)` | ✅ Valid         |
| `def func(*args, **kwargs)`    | ✅ Valid         |
| `def func(**kwargs, *args)`    | ❌ Invalid       |
| `def func(*args, a)`           | ❌ Invalid       |


**Quick Summary**

| Feature             | `*args`       | `**kwargs`      |
| ------------------- | ------------- | --------------- |
| Stores arguments as | Tuple         | Dictionary      |
| Type of arguments   | Positional    | Keyword         |
| Example call        | `func(1,2,3)` | `func(a=1,b=2)` |
| Unpacking symbol    | `*`           | `**`            |


# **7️⃣ Local vs Global Variables**
Local variable

Exists inside the function.

````python
def test():
  x = 10
  print(x)

test()
````

`x` exists only inside the function.

**Global variable**

Defined outside the function.

````python
x = 5

def show():
  print(x)

show()
````

Using `global`

````python
x = 5

def change():
  global x  # Tell Python we want to use the global variable 'x'
  x =10     # Modify the global variable 'x' to 10

print(x)
change()
print(x)
````

# **9️⃣ Lambda Functions**

Lambda functions are anonymous functions.

Syntax
```
lambda parameters: expression
```

````python
square = lambda x: x*x
print(square(5))
````

````python
# Normal function
def square(x):
    return x**2

# Lambda equivalent
square_lambda = lambda x: x**2

print(square(5))
print(f"Using lambda: {square_lambda(5)}")
````

````python
add = lambda a, b: a + b
print(add(3, 4))
````

Example with sorting:

````python
data = [(1,3),(2,1),(4,2)]

data.sort(key=lambda x: x[1])

print(data)
````

````python
data = [(1,3),(2,1),(4,2)]

data.sort(key=lambda x: x[0])

print(data)
````

# **🔟 Higher-Order Functions**

A higher-order function:

1. takes a function as argument
2. returns a function

Example:

````python
def apply_function(func, value):
    return func(value)

def square(x):
    return x*x

print(apply_function(square,5))
````

Common Higher-Order Functions

Python provides built-in ones:
```
map()
filter()
sorted()
```
Example `map`

````python
numbers = [1,2,3]

result = list(map(lambda x: x*2, numbers))

print(result)
````

Example `filter`

````python
numbers = [1,2,3,4]

even = list(filter(lambda x: x%2==0, numbers))

print(even)
````

````python
numbers = [1,2,3,4,5]

result = filter(lambda x: x > 3, numbers)

print(list(result))
````

Example `sorted`

````python
numbers = [5,2,9,1]

print(sorted(numbers))
````

````python
words = ["apple","banana","kiwi"]

print(sorted(words, key=len))
````

***Higher-Order Functions in Python.***

| Function   | Purpose                                            | Syntax                           | Returns                  | Example                            |
| ---------- | -------------------------------------------------- | -------------------------------- | ------------------------ | ---------------------------------- |
| `map()`    | Applies a function to every element in an iterable | `map(function, iterable)`        | Map object (iterator)    | `map(lambda x: x*2, [1,2,3])`      |
| `filter()` | Filters elements based on a condition              | `filter(function, iterable)`     | Filter object (iterator) | `filter(lambda x: x>2, [1,2,3,4])` |
| `sorted()` | Sorts elements using an optional key function      | `sorted(iterable, key=function)` | New sorted list          | `sorted([3,1,2])`                  |

# **1️⃣1️⃣ Recursion**

Recursion = function calling itself.

Example: factorial
```
5! = 5 × 4 × 3 × 2 × 1
```
Recursive solution:

````python
def factorial(n):

    if n == 1:
        return 1

    return n * factorial(n-1)

print(factorial(5))
````

Important parts:
```
base case
recursive case
```

# **1️⃣3️⃣ Decorators**

Decorators modify the behavior of functions.

Example:

````python
def my_decorator(func):

    def wrapper():
        print("Before function")

        func()

        print("After function")

    return wrapper
````

Use decorator:

````python
@my_decorator
def say_hello():
    print("Hello")

say_hello()
````

Decorators are used heavily in:

- web frameworks

- logging

- authentication

- timing functions

Example Real Decorator

Timing a function:

````python
import time

def timer(func):
    def wrapper():
        start = time.time()   # record start time
        func()                # run the original function
        end = time.time()     # record end time
        print("Execution time:", end - start)
    return wrapper
````

````python
@timer
def my_task():
    # Example: simulate a task that takes time
    for i in range(1000000):
        pass
my_task()
````

(depending on how long the loop takes).
