# **Example 1 — Even or Odd Numbers**

Concepts used:

- input

- variables

- if

- modulo

- loop

```python
for num in range(1, 11):
  if num % 2 == 0:
    print(num, "is even")
  else:
    print(num, "is odd")
```

# **Example 2 — Sum of Numbers**

Concepts:

- variables

- loop

- arithmetic

```python
total = 0

for i in range (1, 6):
  #total += i
  total = i + total
print("Total", total)
```

# **Example 3 — Number Guessing**

Concepts:

- input

- while loop

- if condition

```python
secret = 7
guess = 0

while guess != secret:

  guess = int(input("Guess the number: "))

  if guess == secret:
    print("correct")
  else:
    print("try again")
```

# **Example 4 — Multiplication Table**

Concepts:

- nested loops

- arithmetic

```python
for i in range (1, 6):
  for j in range (1, 6):
    print(i * j, end = " ")
  print()
```

# **Example 5 — Find Largest Number**

Concepts:

input

condition

loop

variable tracking

```python
largest = 0

x = int(input("enter num of x: "))

for i in range (x):

  num = int(input("Enter number: "))

  if num > largest:
    largest = num

print("Largest number:", largest)
```

# **Example 6 — Count Even Numbers**

Concepts:

- loop

- condition

- counter

```python
count = 0

for i in range(1, 21):

  if i % 2 == 0:
    count +=1

print("Total even numbers:", count)
```

# **Example 7 — Password Check**

Concepts:

- while

- input

- condition

```python
password = "python123"
user_input = ""

while user_input != password:

   user_input = input("Enter password: ")

   if user_input != password:
    print("Wrong password")

print("Access granted")
```

# **Example 8 — Break Example**

```python
for i in range(10):

  if i == 5:
    break
  print(i)
```

# **Example 9 — Continue Example**

```python
for i in range(10):

    if i % 2 == 0:
        continue

    print(i)
```

# **Example 10 — Pattern Printing**

```python
row = 5

for i in range(1, row+1):
  print("*" * i)
```
